```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")

plt.xlabel("x")
plt.ylabel("y")
plt.title(" sin & cos ")

plt.legend()
plt.show()
```

# 第2章 感知机

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta:
        return 0
    else:
        return 1
```

```python
import numpy as np
x = np.array([0, 1])
w = np.array([0.5, 0.5])

b = -0.7

tmp = w * x

print(np.sum(tmp) + b)
```

# 第3章 神经网络

```python
import numpy as np
x = np.array([-1.0, 1.0, 2.0])
y = x > 0
print(y)
print(type(y))

z = y.astype(np.int)
print(z)
```

```python
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x>0, dtype = np.int)

x = np.arange(-0.5, 5.0, 0.1)
y = step_function(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1) # 指定y轴的范围
plt.show()
```

```python
def sigmoid(x):
    return 1 / ( 1+ np.exp(-x))

# x = np.array([-1.0, 1.0, 2.0])
x = np.arange(-5.0,5.0,0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
x = np.arange(-5.0,5.0,0.1)
y1 = sigmoid(x)
y2 = step_function(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0,5.0,0.1)
y = relu(x)

plt.plot(x,y)
plt.show()
```
